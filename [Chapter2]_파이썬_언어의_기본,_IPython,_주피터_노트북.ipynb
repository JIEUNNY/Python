## **2.3 파이썬 기초**

### **2.3.1 시맨틱**

**들여쓰기:** 파이썬은 공백 문자를 이용해 코드를 구조화하고, 필요한 경우 한 문장에서 세미콜론을 이용하여 여러 문장을 구분한다. 

**모든 것은 객체:** 파이썬을 구성하는 모든 것은 객체 모델이며 일관성을 가진다. 연관된 자료형을 하나의 객체로 간주한다. 


> (참고) https://wikidocs.net/20457

**함수와 객체 메서드 호출:** 함수는 이름을 통해 함수를 사용할 수 있고, 함수의 값을 변수에 대입할 수 있다. 반면 메소드는 객체와 연관되어 사용되며 사숑하고자 하는 대상이 .으로 연결되어있어야한다. (ex) .append())


> (참고) https://velog.io/@yejin20/Python-%ED%95%A8%EC%88%98%EC%99%80-%EB%A9%94%EC%86%8C%EB%93%9C%EC%9D%98-%EC%B0%A8%EC%9D%B4%EC%A0%90

**변수와 인자 전달:** 바운드 변수와 원본 변수는 한 객체에 대해 다른 참조를 하며 같은 객체를 가리킨다. (deepcopy()를 이용할 경우 영향받지 않는다) 

**동적 참조와 강한 타입:** 타입이 달라도 객체 참조가 가능하다. 모든 객체는 특정한 자료형을 가지며 직관적으로 변환을 수행할만한 부분에 변환이 적용되는 강한 타입의 언어다. 

**속성과 메서드:** obj.attribute_name을 통해 속성과 메서드 접근이 가능하다. 

**덕 타이핑:** 객체가 지원하는 메서드나 행동을 알고 싶을 경우 사용한다. 

**모듈 임포트:** 파이썬 파일에서 정의된 함수를 import 하여 불러올 수 있으며 모듈을 다른 이름으로도 import 가능하다.

### **2.3.2 스칼라형**

**숫자 자료형:** int, float형이 있으며 파이썬에서는 별도의 double 자료형이 존재하지 않는다. 

**문자열:** str 자료형이며 유니코드 문자열이다. 파이썬의 문자열은 변경 불가능하다. 또한 여러 객체는 str 함수를 통해 문자열로 변환할 수 있다. 문자열 옆에 r을 사용하면 원 형태 그대로 해석할 수 있다. 

1) {0:.2f} 첫번째 인자를 소수점 아래 2자리까지만 표시한다.

2) {1:s} 두번째 인자가 문자열이다. 

3) {2:d} 세번째 인자가 정수다. 

**바이트와 유니코드:** 유니코드가 최상위 문자열 타입이다. 타입 지정을 통해 utf-8로 인코딩할 수 있으며 원 형태로 되돌릴 수 있다. 

**불리언:** 비교 조건 식의 결과로 true, false이다. 

**형변환:** str, bool, int, float는 자료형 변환을 위해 사용된다. 

**None:** null 값을 의미한다. 

```
a = 'this is programming seminar'
a[10] = 'f'

# 값을 대체할 수 없으며 에러가 발생하는 것을 확인할 수 있다. 
```

```
b = a.replace('programming', 'winter programming')
b

# 이때 a 값은 변하지 않는다. 
```

```
# 자료형 형변환
s = '3.14159'
fval = float(s)
type(fval)
```

**날짜와 시간**

```
from datetime import datetime, date, time
dt = datetime(2021, 1, 13, 11, 59, 59)
dt.day
```

```
dt.minute
```

```
# date와 time 메서드를 이용해 날짜와 시간을 추출할 수 있다.
dt.date()
dt.time()
```

```
# strftime 메서드로 datetime의 타입을 문자열로 만들 수 있다. 
dt.strftime('%m/%d/%Y %H:%M')
```

```
# strptime 메서드로 datetime의 타입을 객체로 만들 수 있다.
datetime.strptime('20210113', '%Y%m%d')
```

### **2.3.3 흐름 제어**

**if, elif, else:** 조건을 검사하고 수행할 부분을 정의할 수 있다. 

**for 문:** continue등을 통해서 실행 여부를 조정할 수 있고, 범위 내에서 반복을 수행한다. 

**while 문:** 조건을 명시하여 조건이 부합하지 않거나 종료 조건이 있을 경우 반복을 중단한다. 

**range:** 지정한 특정 범위 내에서 이터레이터를 반환한다. 

**삼항 표현식:** 여러 문장의 코드를 한 문장으로 표현할 수 있게 한다. 

```
# 삼항 표현식 

x = 5
'Non-negative' if x >= 0 else 'Negative'
```
